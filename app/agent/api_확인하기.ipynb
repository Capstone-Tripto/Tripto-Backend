```python
# 원하는 정보만 필터링
import requests
import json

def get_tourism_data(lDongRegnCd, content_type, lDongSignguCd=None):
    key = "3ee851822ac974d2b7d447a4d6da1eac83d489b1a0e0e647c56af6aa602f60e5"
    url = "https://apis.data.go.kr/B551011/KorService2/areaBasedList2"

    params = {
        'serviceKey': key,
        'numOfRows': '10',
        'pageNo': '1',
        'MobileOS': 'ETC',
        'MobileApp': 'Tripto',
        '_type': 'json',
        'arrange': 'A',                # A=제목순 
        'lDongRegnCd' : lDongRegnCd,   # 법정동 시도 코드
        'contentTypeId': content_type  # 12: 관광지, 32: 숙박, 39: 음식점 
    }

    # 있으면 파라미터 추가
    if lDongSignguCd:
        params['lDongSignguCd'] = lDongSignguCd

    response = requests.get(url, params=params)
    return response.json()



code_list = [
    {"name":"서울", "code":"11"},
    {"name":"부산", "code":"26"},
    {"name":"김포", "code":"41","code_2":"570"},
]


def find_code(city_name):
    # 도시명으로 검색하여 (시도코드, 시군구코드) 튜플 반환
    for item in code_list:
        if item['name'] == city_name:
            return item.get('code'), item.get('code_2')
    return None, None

city_name = input("어느 지역을 원하시나요?")
code, code_2 = find_code(city_name)

# 1. 관광지(12) 정보 가져오기
tourist_spots = get_tourism_data(code, '12', code_2)
print("--- 관광지 목록 ---")
print(tourist_spots)

# 2. 음식점(39) 정보 가져오기
restaurants = get_tourism_data(code, '39', code_2)
print("\n--- 음식점 목록 ---")
print(restaurants)

# 3. 숙박(32) 가져오기
Accommodation = get_tourism_data(code, '32', code_2)
print("\n--- 숙박 목록 ---")
print(Accommodation)
```

```python
# 반복 정보 조회

def search (id, type):
    key = "3ee851822ac974d2b7d447a4d6da1eac83d489b1a0e0e647c56af6aa602f60e5"
    url = "http://apis.data.go.kr/B551011/KorService2/detailInfo2"

    # 요청 파라미터
    params = {
        'serviceKey': key,
        'numOfRows': '10',
        'pageNo': '1',
        'MobileOS': 'ETC',
        'MobileApp': 'Tripto',
        '_type': 'json',
        'contentId' : id,   
        'contentTypeId': type  # 12: 관광지, 32: 숙박, 39: 음식점
    }
    
    response = requests.get(url, params=params)
    return response.json()


all_details = []

for id, type in zip(df["contentid"], df["contenttypeid"]):
    print("조회중")
    reault = search (id, type)

    # 데이터가 정상적으로 왔는지 확인 후 리스트에 추가
    if reault and 'response' in reault and 'body' in reault['response']:
        items = reault['response']['body']['items']
        if items: # 데이터가 있는 경우만
            # 반복 정보는 한 ID당 여러 개(item 리스트)일 수 있으므로 extend 사용
            all_details.extend(items['item'])
```
